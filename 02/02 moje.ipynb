```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt', sep=' ', header=None, names=['opponent','me'])
df
```

```python
# points for particular "round entry"
mustra = {
    'X' : 1,
    'Y' : 2,
    'Z' : 3
}
```

```python
df['me_points']=df['me'].map(mustra)
df
```

```python
df['game']=df['opponent']+df['me']
df
```

```python
possible_results = {
    'AX': 3 ,
    'AY': 6 ,
    'AZ': 0 ,
    'BX': 0 ,
    'BY': 3 ,
    'BZ': 6 ,
    'CX': 6 ,
    'CY': 0 ,
    'CZ': 3 ,
    }
```

```python
df['game_points']=df['game'].map(possible_results)
df
```

```python
result = df['me_points'].sum()+df['game_points'].sum()
print(f'Part A result:\n{result}')
```

# part B:

```python
df = pd.read_csv('input.txt', sep=' ', header=None, names=['opponent','game_result'])
df
```

```python
g_points = {
    'X' : 0,
    'Y' : 3,
    'Z' : 6
}
```

```python
df['game_points']=df['game_result'].map(g_points)
df
```

```python
df['tg']=df['opponent']+df['game_result']
df
```

```python
mustra_points = {
    'A' : 1,
    'B' : 2,
    'C' : 3
}
```

```python
my_turns = {
    'AX': 'C' ,
    'AY': 'A' ,
    'AZ': 'B' ,

    'BX': 'A' ,
    'BY': 'B' ,
    'BZ': 'C' ,

    'CX': 'B' ,
    'CY': 'C' ,
    'CZ': 'A' ,
    }
```

```python
df['m_turns']=df['tg'].map(my_turns)
df
```

```python
df['m_turns_points']=df['m_turns'].map(mustra_points)
df
```

```python
result = df['game_points'].sum()+df['m_turns_points'].sum()
print(f'Part B result:\n{result}')
```
