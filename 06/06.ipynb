```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt', header=None)
df
```

```python
df.loc[0,0]
```

```python
mystring = df.loc[0,0]
mystring
```

```python
# maybe simplest way would be just:
# mystring = open('input.txt').read()
# mystring
```

# Part A:

```python
newlist = []
for i in range(len(mystring)-4):
    newlist.append(mystring[i:i+4])
newlist
```

```python
new_df=pd.DataFrame(newlist)
new_df
```

```python
set('czfz')
```

```python
new_df[1] = new_df[0].apply(set)
new_df
```

```python
new_df[1].agg(len)
```

```python
new_df[2] = new_df[1].apply(lambda x:len(x))
new_df
```

```python
new_df[(new_df[2]==4)]
```

```python
new_df[(new_df[2]==4)].head(1)
```

```python
new_df[(new_df[2]==4)].head(1).index[0]
```

```python
new_df[2].idxmax() # aj toto ide
```

```python
# we get the index position, but need to add 4 (after which the "signal" occurs)
result = new_df[(new_df[2]==4)].index[0] + 4 
print(f'Part A result:\n{result}')
```
