```python
df.ne('').sum().sum()
```

```python
n_crates = (df!='').sum().sum()
# n_crates = df.ne('').sum().sum()
# n_crates = df.applymap(lambda x:True if x!='' else False).sum().sum()
# n_crates = df.notna().sum().sum() # 56 crates altogether
n_crates
```

```python
df = df.reindex(range(n_crates), fill_value='')
df
```

```python
df_saved = df.copy()
```

```python
# read moves
df_moves = pd.read_csv('input.txt', header=None, skiprows=10, names=['moves'])
df_moves
```

```python
# toto bude regex:
# r'(\d+) from (\d) to (\d)'
```

```python
# extract moves
df_moves = df_moves['moves'].str.extract(r'(\d+) from (\d) to (\d)') 
df_moves = df_moves.applymap(int)
df_moves
```

```python
def engine_func(ser, crate_order):
    amount_ , from_ , to_ = ser
    df[to_] = df[to_].shift(amount_)
    df.loc[ :amount_-1 , to_] = df.loc[ :amount_-1 , from_][::crate_order].values
    # df.iloc[0:amount_ , to_-1] = df.iloc[ 0:amount_ , from_-1][::-1]
    df[from_] = df[from_].shift(-amount_ , fill_value='')
```

```python
df_moves.apply( engine_func , axis=1, crate_order=-1)
```

```python
df # final state
```

```python
result = df.loc[ 0, :].sum()
print(f'Part A result:\n{result}')
```
