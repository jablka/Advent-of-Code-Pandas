```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt', header=None, names=['elf1', 'elf2'])
df
```

```python
df['elf1_'] = df['elf1'].str.split('-')
df['elf2_'] = df['elf2'].str.split('-')
df
```

```python
df['elf1__'] = df['elf1_'].apply(lambda x:set( range(int(x[0]),int(x[1])+1) ) )
df['elf2__'] = df['elf2_'].apply(lambda x:set( range(int(x[0]),int(x[1])+1) ) )
df
```

```python
df['calculate'] = df.apply(lambda x:(x['elf1__'].issubset(x['elf2__']) or x['elf2__'].issubset(x['elf1__']) ) , axis=1)
df
```

```python
result = df['calculate'].sum()
print(f'Part A result:\n{result}')
```

# part2:

```python
df['calculatePrienik'] = df.apply(lambda x:True if x['elf1__'].intersection(x['elf2__']) else False , axis=1)
df
```

```python
result = df['calculatePrienik'].sum()
print(f'Part B result:\n{result}')
```
